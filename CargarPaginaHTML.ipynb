###**Leyendo Banco de datos**
Creando un banco de datos local

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
import pandas as pd
```

```python
archivo = '/content/clientes_banco.csv'
```

```python
datos = pd.read_csv(archivo)
```

```python
datos.head()
```

```python
datos.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

##**Borrando Registros**

```python
from sqlalchemy.exc import SQLAlchemyError
delete = "delete from clientes where ID_Cliente = 5008804"
try:
  r_set = engine.connect().execute(text(delete))
except SQLAlchemyError as e:
  #error = str(e.__dict__['orig'])
  print(e)
else:
  print("#Registros Borrados", r_set.rowcount)
```

```python
pd.read_sql_table('clientes', engine)
```

###**Update Registro**

```python
update = "update clientes set Grado_estudio = 'Nivel superior' where ID_Cliente = 5008808"
try:
  r_set = engine.connect().execute(text(update))
except SQLAlchemyError as e:
  #error = str(e.__dict__['orig'])
  print(e)
else:
  print("#Registros Updated", r_set.rowcount)
```

```python
pd.read_sql_table('clientes', engine)
```
